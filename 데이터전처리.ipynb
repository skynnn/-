* pop_cd : cd 고유번호 L011: 지역번호(01:서울, 1:10대, 022:인천20대) Y:성별번호 Y80(80대 남자) 
* year : 년도 
* month : 월 
* avg_score : 신용점수  
* avg_rat:  신용등급 
* city : 도시  
* sex : 성별 
* ages : 나이대 
* population : pop_cd로 구분된 그룹에 포함된 고객인원 수/경제활동 인구수 
* num_opencard : 카드개설 개수 
* num_usecard : 사용카드 개수 
* monthly_card_spend :  카드 소비 금액 
* monthly_lc : 대출 약정 금액  
* monthly_loan : 대출 금액 합계  
* monthly_bk_loan : 은행대출 금액  
* monthly_cd_loan : 카드대출 금액   
* monthly_installments_loan : 할부 대출 금액 
* monthly_insurance_loan : 보험 대출 금액 
* monthly_sbk_loan : 저축은행 대출 금액 
* loan_commitment : 한도 대출 금액 (마이너스 통장) 
* inst_rep_loanb : 분할 상환 대출 금액  
* ls_rep_loanb : 일시 상환 대출 금액 
* credit_loan : 신용 대출 금액 
* mortgage_loan : 담보 대출 금액 
* credit_card_payment : 신용카드 사용 금액 
* credit_card_installments_payment : 신용카드 할부 사용 금액

```python
df = pd.read_csv('./dataset/credit_card_data.csv', encoding='utf-8')
df
```

##### age 범위 선정

```python
df1 = df.replace(0, np.NaN)
df1.isna().sum()
```

```python
print(df1[df1['monthly_cd_loan'].isna()]['ages'].unique())
print(df1[df1['monthly_installments_loan'].isna()]['ages'].unique())
print(df1[df1['monthly_insurance_loan'].isna()]['ages'].unique())
print(df1[df1['monthly_sbk_loan'].isna()]['ages'].unique())
print(df1[df1['loan_commitment'].isna()]['ages'].unique())
print(df1[df1['inst_rep_loanb'].isna()]['ages'].unique())
print(df1[df1['credit_loan'].isna()]['ages'].unique())
print(df1[df1['credit_card_payment'].isna()]['ages'].unique())
print(df1[df1['credit_card_installments_payment'].isna()]['ages'].unique())
```

10대, 90대의 0인 컬럼이 많고, 대출이라는 특성상 주요타겟층이 아니라고 판단하여

10대, 90대를 제외하고 분석

```python
df = df.drop(df[(df['ages'] == '10대') | (df['ages'] == '90대')].index)
df = df.reset_index(drop=True)
```

##### 변수 추가 1) 새로운 신용등급 계산 (20%비율로)

```python
score_max = df['avg_score'].max()+1
score_20 = df['avg_score'].quantile(0.2)
score_40 = df['avg_score'].quantile(0.4)
score_60 = df['avg_score'].quantile(0.6)
score_80 = df['avg_score'].quantile(0.8)
bins = [score_max, score_80, score_60, score_40, score_20]
df['new_avg_rat'] = np.digitize(df['avg_score'], bins)
df
```

##### 변수 추가 2) 체크카드 사용량

월 카드 총 이용금액(monthly_card_spend) –(신용카드 일시불 이용금액(credit_card_payment) - 신용카드 할부 이용금액(credit_card_installments_payment))

```python
df['debit_card_payment'] = df['monthly_card_spend'] - df['credit_card_payment'] - df['credit_card_installments_payment']
df
```

```python
df['debit_card_payment'].describe()
```

##### 변수 추가 3) 한 달 이내에 사용 가능한 소비량 (20%씩 범주화)

소비량 = 신용카드 일시불 사용량 + 체크카드 사용량

```python
df['monthly_spend'] = df['credit_card_payment'] + df['debit_card_payment']
df
```

```python
df['monthly_spend'].describe()
```

```python
# 20% 비율로 범주화
spend_max = df['monthly_spend'].max()+1
spend_20 = df['monthly_spend'].quantile(0.2)
spend_40 = df['monthly_spend'].quantile(0.4)
spend_60 = df['monthly_spend'].quantile(0.6)
spend_80 = df['monthly_spend'].quantile(0.8)
bins = [spend_max, spend_80, spend_60, spend_40, spend_20]
df['monthly_spend_rat'] = np.digitize(df['monthly_spend'], bins)
df
```

```python
df.groupby('monthly_spend_rat').count()['pop_cd']
```

```python
# 컬럼 순서 변경
df = df[['pop_cd', 'year', 'month', 'avg_score', 'avg_rat', 'new_avg_rat', 'city', 'sex',
       'ages', 'monthly_spend_rat', 'population', 'num_opencard', 'num_usecard',
       'monthly_card_spend', 'monthly_lc', 'monthly_loan', 'monthly_bk_loan',
       'monthly_cd_loan', 'monthly_installments_loan',
       'monthly_insurance_loan', 'monthly_sbk_loan', 'loan_commitment',
       'inst_rep_loanb', 'ls_rep_loanb', 'credit_loan', 'mortgage_loan',
       'credit_card_payment', 'credit_card_installments_payment',
       'debit_card_payment', 'monthly_spend']]
```

##### 컬럼 * population

```python
df_total = df.iloc[:, :11]
for col in df.columns[3:]:
    if type(df['{}'.format(col)][0]) in [np.int64, np.float64, np.int32] and col!='population':
        colname = 'total_' + col
        df_total['{}'.format(colname)] = df['{}'.format(col)] * df.population
df_total.head(3)
```

```python
# 시계열을 위해 '년-월-일' 형식으로 변경, 매월 1일로 임의 설정
df['year'] = df['year'].apply(str)
df['month'] = df['month'].apply(str)
df['date'] = df['year'] + ' ,' + df['month'] + ' ,1'
df['date'] = pd.to_datetime(df['date'])

df_total['year'] = df_total['year'].apply(str)
df_total['month'] = df_total['month'].apply(str)
df_total['date'] = df_total['year'] + ' ,' + df_total['month'] + ' ,1'
df_total['date'] = pd.to_datetime(df_total['date'])
```

도시별, 성별 데이터프레임 분리

```python
# 도시별, 성별 데이터프레임 분리
city_ages = df[df['pop_cd'].str.startswith('L')]
sex_ages = df[df['pop_cd'].str.startswith('Y')]
# 필요없는 컬럼 삭제
city_ages = city_ages.drop('sex', axis=1)
sex_ages = sex_ages.drop('city', axis=1)

# 도시별, 성별 데이터프레임 분리
city_ages_total = df_total[df_total['pop_cd'].str.startswith('L')]
sex_ages_total = df_total[df_total['pop_cd'].str.startswith('Y')]
# 필요없는 컬럼 삭제
city_ages_total = city_ages_total.drop('sex', axis=1)
sex_ages_total = sex_ages_total.drop('city', axis=1)
```

##### dataset -> csv 저장

```python
df.to_csv('./dataset/credit_card_dataset.csv', encoding='utf-8', index=False)
df_total.to_csv('./dataset/credit_card_dataset_total.csv', encoding='utf-8', index=False)
city_ages.to_csv('./dataset/credit_card_dataset_city.csv', encoding='utf-8', index=False)
sex_ages.to_csv('./dataset/credit_card_dataset_sex.csv', encoding='utf-8', index=False)
city_ages_total.to_csv('./dataset/credit_card_dataset_city_total.csv', encoding='utf-8', index=False)
sex_ages_total.to_csv('./dataset/credit_card_dataset_sex_total.csv', encoding='utf-8', index=False)
```
